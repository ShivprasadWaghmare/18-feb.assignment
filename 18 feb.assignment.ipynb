```python
Q1. What is an API? Give an example, where an API is used in real life.
ans:
     an API can be used to connect virtually any processes.
     e.g.
     A.Sharing flight information between airlines and travel sites. 
     B.Using Google Maps in a rideshare app.
     C.Online Payment Processing
     
     
        
Q2. Give advantages and disadvantages of using API.
ANS:

    ADVANTAGES   of APIs:
Faster development: Leverage existing functionality, saving time and effort.
Increased functionality: Add features to your app without building them yourself.
Improved user experience: Create seamless integrations with other apps.
Scalability: Build applications that can handle growing demand easily.
New revenue streams: Monetize your data or functionality through paid APIs.


    Disadvantages of APIs:
Security risks: APIs can introduce vulnerabilities if not properly secured.
Dependency: Reliant on the stability and availability of the external API.
Complexity: Learning and using APIs can add complexity to your project.
Cost: Some APIs require paid subscriptions or usage fees.
Vendor lock-in: Switching to another API can be difficult and expensive


Q3. What is a Web API? Differentiate between API and Web API.
ans:Web API:
Definition: A web-based API designed for interaction over the internet using HTTP protocol.

Access: Accessed through URLs and interacts with data formats like JSON and XML.

Purpose: Facilitates communication between web servers and clients, enabling data exchange and functionality sharing.

Examples: Retrieving weather data, posting social media updates, integrating payment gateways.


Feature                                	API	Web                                                                  API

1.Scope	              Broader, encompasses various types of interfaces	                                     Specific to web-based interactions


2.Purpose	 Wide range of uses, including software integration and system interactions	       Primarily for web-based communication and data exchange

3.Protocol	      Can use diverse protocols (HTTP, FTP, SOAP, etc.)	                                  Primarily leverages HTTP

4.Data format	                  Varied formats	                                                     Often JSON or XML



Q4. Explain REST and SOAP Architecture. Mention shortcomings of SOAP.
ans:

    REST:

Architectural style for APIs, based on web principles like HTTP verbs and resource identifiers.
Pros: Flexible, lightweight, stateless, scalable, supports various data formats (JSON, XML etc.).
Cons: Less strict, requires more design effort, limited security features.

    SOAP:

Protocol for accessing web services, uses XML messages and envelopes.
Pros: Standardized, secure, reliable, good for complex transactions.
Cons: Verbose, heavy messages, less scalable, limited flexibility, only XML data format.


    SOAP Shortcomings:
Verbosity: Large XML messages increase bandwidth and processing time.
Complexity: Strict standards make development and implementation complex.
Limited Scalability: Stateful nature and heavy messages restrict scaling.
Flexibility: Less adaptable to diverse data formats and platforms.
  
  
 Q5. Differentiate between REST and SOAP.
 ans:

Structure:

REST: Architectural style, flexible, no strict rules.
SOAP: Protocol, rigid, follows specific message format (XML).

Data format:

REST: Supports various formats like JSON, XML, plain text.
SOAP: Uses XML only.

Performance:

REST: Faster, smaller messages, easier caching.
SOAP: Slower, larger messages, complex processing.

State:

REST: Stateless, each request independent.
SOAP: Stateful, server maintains context of previous requests.

Security:

REST: Requires additional security measures.
SOAP: Built-in security features like WS-Security.

Complexity:

REST: Easier to understand and implement.
SOAP: More complex syntax and structure.

Use cases:

REST: Web APIs, lightweight communication, modern apps.
SOAP: Legacy systems, secure transactions, enterprise integration.

Basically, REST is simpler, faster, and more flexible, while SOAP is stricter, more secure, and better for complex transactions. Choosing between 
them depends on your specific needs.











```
